```python
import coral_lib.patch.tools as cpt
import catlin_tools
workdir = '/runs/catlin/car/assess_scale1.0/'
gpuid = 0
net = bct.load_model(workdir, 'snapshot_iter_16000.caffemodel', net_prototxt = 'deploy.prototxt', gpuid = gpuid)
pyparams = cpt.pyparams_from_net(osp.join(workdir, 'testnet.prototxt'))

labelset = catlin_tools.get_car_labelset()
imlist, imdict = catlin_tools.load_caribbean_data('test', labelset)

f = open(osp.join(workdir, 'testcovers.csv'), 'w')
f.write('imid')
for label in labelset:
    f.write(', ' + str(label))
f.write('\n')
for im in tqdm(imlist):
    (gtlist, estlist, scorelist) = cpt.classify_from_patchlist([im], imdict, pyparams, net)
    cover = np.bincount(estlist, minlength=len(labelset)) / float(len(estlist))
    f.write(osp.basename(im))
    for c in cover:
        f.write(', ' + str(c))
    f.write('\n')
f.close()
```
