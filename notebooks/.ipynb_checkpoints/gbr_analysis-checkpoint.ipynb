```python
## SETUP BUT WITH RANDOM POINT LOCATIONS
#
def make_gbr_testdata(npoints):
    # make imlist
    lines = [line.rstrip('\n') for line in open('/data/catlin/gbr/input_data/test/annotations.csv')]
    imlist = set()
    for line in lines[1:]:
        parts = line.split(',')
        parts = [p.strip() for p in parts]
        imname = parts[0] + '.jpg'
        imlist.add(imname)
    imlist = list(imlist)
    imlist = ['/data/catlin/gbr/input_data/test/images/' + imname for imname in imlist]
    imdict = {}
    for imname in imlist:
        im = np.asarray(Image.open(imname))
        [nrows, ncols, _] = im.shape
        imdict[osp.basename(imname)] = ([], 100)
        rows = np.random.choice(np.linspace(50, nrows - 50, nrows - 99), npoints, replace = False)
        cols = np.random.choice(np.linspace(50, ncols - 50, ncols - 99), npoints, replace = False)
        for row, col in zip(rows, cols):
            imdict[osp.basename(imname)][0].append((int(row), int(col), 0))
    return imlist, imdict

def classify_gbt_testdata(npoints):
    imlist, imdict = make_gbr_testdata(npoints)
    gpuid = 0
    net = bct.load_model('/runs/catlin/gbr/ratio15/', 'snapshot_iter_17000.caffemodel', net_prototxt = 'testnet.prototxt', gpuid = gpuid)
    pyparams = pload('/runs/catlin/gbr/ratio15/pyparams.pkl')
    labelset = pload('/data/catlin/gbr/labelset.p')
    f = open('/runs/catlin/gbr/ratio15/testcovers_{}new.csv'.format(npoints), 'w')
    f.write('imid')
    for label in labelset:
        f.write(', ' + str(label))
    f.write('\n')
    for im in tqdm(imlist):
        (gtlist, estlist, scorelist) = bct.classify_from_patchlist([im], imdict, pyparams, net)
        cover = np.bincount(estlist, minlength=len(labelset)) / float(len(estlist))
        f.write(osp.basename(im))
        for c in cover:
            f.write(', ' + str(c))
        f.write('\n')
    f.close()

classify_gbt_testdata(40)
classify_gbt_testdata(80)
classify_gbt_testdata(160)
```

```python
import catlin_tools
gpuid = 2
npoints = 50
net = bct.load_model('/runs/catlin/gbr/scale1/', 'snapshot_iter_19000.caffemodel', net_prototxt = 'testnet.prototxt', gpuid = gpuid)
net.blobs['label'].reshape(npoints, 1)
net.blobs['data'].reshape(npoints, 3, 224, 224)
net.blobs['label_data_1_split_0'].reshape(npoints, 1)
net.blobs['label_data_1_split_1'].reshape(npoints, 1)
pyparams = pload('/runs/catlin/gbr/scale1/pyparams.pkl')
labelset = pload('/data/catlin/gbr/labelset.p')

indir_root = '/data/catlin/gbr/server_data/'
for indir in glob.glob(indir_root + 'exp*'):
    outdir = indir + '/coverages'
    imgs = glob.glob(indir + '/images/*.jpg')
    print "Starting", indir, len(imgs)
    for img in imgs[:1]:
        imname = osp.basename(img)
        covname = osp.join(outdir, imname + '.cover.csv')
        if osp.isfile(covname):
            continue
        estlist = catlin_tools.classify_image(img, pyparams, npoints, net)
        cover = np.bincount(estlist, minlength=len(labelset)) / float(len(estlist))
        f = open(covname, 'w')
        f.write(imname)
        for c in cover:
            f.write(', ' + str(c))
        f.write('\n')
        f.close()
    print "Done", indir, len(imgs)
```
